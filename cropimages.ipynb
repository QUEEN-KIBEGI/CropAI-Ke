```python

# Define root folder of downloaded images
root_dir = "/content"

# Prepare a list to store records
data = []

# Traverse through each subfolder
for folder in os.listdir(root_dir):
    folder_path = os.path.join(root_dir, folder)
    if os.path.isdir(folder_path):
        # Check if the folder is one of the image category folders
        if folder in ["Healthy_Tomato", "Healthy_Maize", "Healthy_Potato", "Healthy_Wheat", "Healthy_Banana", "Tomato_Blight", "Maize_Leaf_Spot", "Potato_Early_Blight", "Wheat_Rust", "Banana_Bacterial_Wilt"]:
            # Assign label
            label = "healthy" if folder.lower().startswith("healthy") else "unhealthy"

            # Collect image file paths
            for filename in os.listdir(folder_path):
                if filename.lower().endswith((".jpg", ".jpeg", ".png")):
                    full_path = os.path.join(folder_path, filename)
                    data.append({"image_path": full_path, "label": label})

# Create DataFrame and save to CSV
df = pd.DataFrame(data)
df.to_csv("plant_disease_dataset.csv", index=False)

print("✅ CSV created: plant_disease_dataset.csv")
```

## Step 7: Data Modeling

- Loading of pre-trained base model (MobileNetV2)
- Freezing of base layers
- Building of classifier
- Model Compilation
- Model Training


```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras import layers, models
from tensorflow.keras.losses import BinaryFocalCrossentropy
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.model_selection import train_test_split
from sklearn.utils import class_weight
import numpy as np
import os

# === Split Data ===
train_df, test_df = train_test_split(df, test_size=0.2, stratify=df['label'], random_state=42)

# === Image Size & Batch ===
IMG_SIZE = 224
BATCH_SIZE = 32

# === Data Augmentation (Stronger) ===
augmentor = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.15,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(rescale=1./255)

train_data = augmentor.flow_from_dataframe(
    train_df,
    x_col='image_path',
    y_col='label',
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=True
)

val_data = val_datagen.flow_from_dataframe(
    test_df,
    x_col='image_path',
    y_col='label',
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=False
)

# === Compute Class Weights (Optional but Useful) ===
class_weights = class_weight.compute_class_weight(
    class_weight='balanced',
    classes=np.unique(train_df['label']),
    y=train_df['label']
)
class_weights = dict(enumerate(class_weights))

# === Load Pretrained Base ===
base_model = MobileNetV2(include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3), weights='imagenet')
base_model.trainable = False

# === Build Model ===
model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.BatchNormalization(),               # ✅ Added
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.5),                        # ✅ Increased
    layers.Dense(1, activation='sigmoid')
])

# === Focal Loss with Label Smoothing ===
focal_loss = BinaryFocalCrossentropy(gamma=2.0, label_smoothing=0.1)

# === Callbacks ===
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, min_lr=1e-6)

# === Compile & Train (Frozen Base) ===
model.compile(optimizer='adam', loss=focal_loss, metrics=['accuracy'])

history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=10,
    class_weight=class_weights,
    callbacks=[early_stop, reduce_lr]
)

# === Fine-tune Last 50 Layers ===
base_model.trainable = True
for layer in base_model.layers[:-50]:
    layer.trainable = False

# === Re-Compile with Lower LR ===
model.compile(optimizer=tf.keras.optimizers.Adam(1e-5), loss=focal_loss, metrics=['accuracy'])

history_finetune = model.fit(
    train_data,
    validation_data=val_data,
    epochs=10,
    class_weight=class_weights,
    callbacks=[early_stop, reduce_lr]
)
```

## Step 8: Model Evaluation

```python
# Evaluate the model
loss, accuracy = model.evaluate(val_data)

print(f"\n✅ Test Loss: {loss:.4f}")
print(f"✅ Test Accuracy: {accuracy:.4f}")
```

## Step 9: Visualisation of model performance

```python
# Plot training history
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()

plt.tight_layout()
plt.show()
```

## Step 10: Data Frame of Model Predictions

```python
import numpy as np

# Make predictions on the validation data
predictions = model.predict(val_data)

# Convert predictions to binary labels (0 or 1)
predicted_labels = (predictions > 0.5).astype(int)

# Get true labels from the validation data generator
true_labels = val_data.classes

# Get the filenames of the validation images
filenames = val_data.filenames

# Create a DataFrame to display predictions and true labels
results_df = pd.DataFrame({
    'filename': filenames,
    'true_label': true_labels,
    'predicted_label': predicted_labels.flatten()
})

# Map numerical labels back to original class names
label_map = {v: k for k, v in val_data.class_indices.items()}
results_df['true_label_name'] = results_df['true_label'].map(label_map)
results_df['predicted_label_name'] = results_df['predicted_label'].map(label_map)

# Display the first few results
print("✅ Sample Predictions:")
display(results_df.head())
```
